LAGRANGIAN AND HAMILTONIAN OF A SYSTEM

here i have assumed generalised co ordinates, and lagrangian or hamiltonian to be not explicitely depended on time, thus the partial derivative w.r.t. time is 0.

```python
#headers
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr
import numpy as np
#import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
```

```python

t = sp.symbols("t")
q_i = sp.Function('q')(t)  # q(t)
#_dot = sp.symbols('_dot')  # q'(t)

co_ordinate = []  # Initializing as an empty NumPy array for co - ordinates
const = []  # Initialize constants as empty array for the constant symbols
```

```python
'''user input for generalised coordinates and constants'''

# Input for coordinates (except time)
while True:
    variables_input = input("Enter the variables one by one (except time or t) or 'exit' when done: ")
    if variables_input == "exit":
        break
    else:
        co_ordinate = np.append(co_ordinate, sp.Function(variables_input)(t))

# Input for constants
while True:
    constant_input = input("Enter the constants one by one (except time or t) or 'exit' when done: ")
    if constant_input == "exit":
        break
    else:
        const = np.append(const, sp.Symbol(constant_input))

# Check the results
print("Coordinates:", co_ordinate)
print("Constants:", const)



```

```python
# Define kinetic energy function
# Build locals once: map q -> q(t), q_dot -> dq/dt
locals_map = {'t': t}
for q_i in co_ordinate:
        # q should be an AppliedUndef like x(t)
        base = q_i.func.__name__      # "x" from x(t)
        locals_map[base] = q_i
        locals_map[f'{base}_dot'] = sp.diff(q_i, t)
        # Optional: higher derivatives
        locals_map[f'{base}_ddot'] = sp.diff(q_i, t, 2)
```

```python

#KE
def get_KE(co_ordinate):
    KE = 0
    while True:
        KE_input = input("Enter kinetic energy term (or 'exit' to finish): ").strip()
        if KE_input.lower() == "exit":
            break
        # Parse once per input using the full locals map
        KE += sp.sympify(KE_input, locals=locals_map)
    return KE


#PE
def get_PE(co_ordinate):
    PE = 0
    while True:
      PE_input = input("Enter PE term (or 'exit'): ")
      if PE_input == "exit":
        break
      PE += sp.sympify(PE_input, locals=locals_map)

      return PE
```

```python

KE = get_KE(co_ordinate)
PE = get_PE(co_ordinate)

print(KE)
print(PE)



# Lagrangian
L = KE - PE
print(f"Lagrangian (L) = {L}")
```

```python
'''getting the lagrangian equation, solution and hamiltonian'''
#q_dot_p=0 # (dq_i/dt) * p
q_dot_p = 0
for q_i in co_ordinate:# for each generalised co-ordinate
    print(q_i)
    q_i_dot=sp.diff(q_i,t)
    print(q_i_dot)
    p_i = sp.diff(L,q_i_dot)#conjugate momentum
    print(f"conjugate momentum: {p_i}")

    #lagrangian equation= {d/dt (dl/dq_i_dot)} - dl/dq_i =0
    equation = sp.diff(p_i,t)  -  sp.diff(L,q_i)
    print(f"Lagrangian equation :{equation}=0")
    sol= sp.dsolve(sp.Eq((equation),0), q_i)
    print(f"the solution: {sol}")
    q_dot_p += q_i_dot * p_i# getting the q_i_dot * p_i term summation over i

    #initial condition
    sol_ini= sp.dsolve(sp.Eq((equation),0), q_i, ics={q_i.subs(t, 0): 0, q_i.diff(t).subs(t, 0): 10})   #put initial conditions for plotting the solutions
    print(f"the solution with initial conditon(0,0): {sol_ini}")
H = sp.Function('H')(t)# initializing Hamiltonian
H= q_dot_p - L #expression for hamiltonian from lagrangian
print(f"Hamiltonian (H) in terms of co- ordinates = {H}")

```
